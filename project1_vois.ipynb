```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("DoctorVisits - DA.csv")
```

```python
df.columns
```

```python

# Display basic information about the dataset
print("Dataset Info:")
print(df.info())
```

```python
# Display summary statistics of numerical columns
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Explore relationships between variables
# Example: Relationship between age and number of doctor visits
age_visits_corr = df['age'].corr(df['visits'])
print("\nCorrelation between age and number of visits:", age_visits_corr)
```

```python
# Example: Average number of visits by gender
avg_visits_by_gender = df.groupby('gender')['visits'].mean()
print("\nAverage number of visits by gender:")
print(avg_visits_by_gender)
```

```python
# Example: Distribution of visits by income level
visits_by_income = df.groupby('income')['visits'].sum()
print("\nDistribution of visits by income level:")
print(visits_by_income)
```

```python
# Example: Frequency of visits for different illness categories
visits_by_illness = df.groupby('illness')['visits'].count()
print("\nFrequency of visits for different illness categories:")
print(visits_by_illness)
```

```python
# Example: Frequency of visits for different illness categories
visits_by_illness = df.groupby('illness')['visits'].count()
print("\nFrequency of visits for different illness categories:")
print(visits_by_illness)
```

```python
# Example 2: Boxplot of visits by gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='visits', data=df, palette='pastel')
plt.title('Visits by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Visits')
plt.show()
```

```python
# Example 3: Bar plot of average visits by income level
plt.figure(figsize=(10, 6))
sns.barplot(x='income', y='visits', data=df, ci=None, palette='viridis')
plt.title('Average Visits by Income Level')
plt.xlabel('Income Level')
plt.ylabel('Average Number of Visits')
plt.show()
```

```python
# Example 4: Count plot of visits by illness category
plt.figure(figsize=(10, 6))
sns.countplot(x='illness', data=df, palette='Set2')
plt.title('Visits by Illness Category')
plt.xlabel('Illness Category')
plt.ylabel('Number of Visits')
plt.xticks(rotation=45)
plt.show()
```
